```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
import statistics 
```

```python
def task1(time, start):
    dist = start
    dist_arr = []
    time_arr = [i for i in range(time)]

    for i in range(time):
        rand = random.randint(0, 2)
        if rand  == 0:
            dist = dist + 1
        elif rand == 1:
            dist = dist - 1
        else:
            pass
        dist_arr.append(dist)

    return (dist_arr, time_arr)

```

```python
dist_arr, time_arr = task1(1000, 0)
plt.plot( time_arr, dist_arr)
plt.show()
```

```python
avg_distarr = []
for i in range(1000):
    dist_arr = []
    time_arr = []
    avg_step = []
    for i in range(1000):
        dist, time = task1(1000, 0)
        dist_arr.append(dist)
        time_arr.append(time)
    for j in range(1000):
        dist_sum = 0
        for i in range(1000):
            dist_sum = dist_sum + dist_arr[i][j]
        avg_step.append(dist_sum/1000)

    avg_distarr.append(sum(avg_step))

print(avg_distarr)
```

```python
sd = statistics.stdev(avg_distarr)
print(sd)
```
